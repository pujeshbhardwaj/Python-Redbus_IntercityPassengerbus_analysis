**In the above result where we obtained correlation coefficient to be -0.637681, that means there is strong negative correlation between bus rating and seats vacant ie. if bus rating decreases then the number of seats left vacant before the journey will increase, and both are inversely propotional to each other  
That means the number of seats booked are dependent on rating of that vehicle.

```python
fig=plt.figure(figsize=(12,5))
ax1=plt.subplot(1,1,1)
ax1.bar(df6.s_no,df6.Seats_Vacant_on_particular_day, color='#6090C0')
ax2=ax1.twinx()
ax2.plot(df6.s_no,df6.Bus_Rating, 'o-', color='#cf6a63', linewidth=3)
plt.title('Combine graph-relation between seat vacant on journy day and Rating', fontsize=20, weight='bold')
plt.show()
```

**In above graph also it can be clearly seen that as the bus rating decreases, the number of seats vacant of that particular bus on journey day increases**

### Results

**Hypothesis1 - 'Bus rating provided by commuters directly related and influence number of seats booked on end day of journey' This assumption is proved to be right in our analysis since there is strong negative correlation between bus rating and seats vacant, where we obtained correlation coefficient to be -0.637681. ie. if bus rating decreases then the number of
seats left vacant before the journey will increase, and both are inversely proportional to each other.**   

**Hypothesis2 - 'Bus rating can be improved by 'MY EICHER' fleet management solutions by resolving problems of commuters on realtime basis' Our second hypothesis related to providing real time solutions to passenges in order to improve negative reviews is not feasible since - most of the journeys are at night and it is not possible to resolve realtime issues regarding factors like - Seat comfort, Staff behaviour, cleanliness with the help of 'MY Eicher' telematics services, It might be helpful in future when world will become more digital and connected.**

**Hypothesis3 - 'MY EICHER' fleet management solutions (Telematics service) can help improving factors and therefore reduce negative reviews which is responsible for lower bus rating' Only factors like Punctuality(Most important factor), Live tracking, A/C other and Driving can be improved with the help of 'My Eicher' telematics services since**

**- Telematics service provides predictive diagnostics which helps prevent breakdown of vehicle and most of the punctuality issues are due to some breakdown mainteinance in vehicle.**

- Eicher Telematics service provide advance gprs technology which accurately tracks and record live locations of vehicles
- A/C and other' factor means complaint due to improper A/C functioning and other electronic device dysfuctioning this can be track on My Eichers telematics via required sensors
- Driving is also a critical factor and one can keep an eye on driver driving behaviour like harsh acceleration, sudden braking etc through telematics device.
- Live tracking and rest stop hygiene - these factors for negative reviews can be improved by traking vehicle accurately and can be done by telematics devise accurate gprs technology.

**So this hypothesis valid for some factors.**
